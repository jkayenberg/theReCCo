```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import summer / winter results

with open("results/results_summer.csv", "r") as csvfile:
    results_summer = pd.read_csv(csvfile)

with open("results/results_winter.csv", "r") as csvfile:
    results_winter = pd.read_csv(csvfile)

results_summer = results_summer.sort_values("Price")

results_winter = results_winter.sort_values("Price")
```

```python
# Plot price / success rate (summer)

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(results_summer["Price"], results_summer["Success rate"], linewidth = 1.0)

x = 666
y = 1.0

plt.scatter(x, y, color='#ff0000', zorder=1)

ax.set_ylabel("Success rate")
ax.set_xlabel("Price [â‚¬]")

plt.xlim([0,3000])

plt.savefig('results/results_summer.png')

plt.show()




```

```python
# Plot price / success rate (winter)

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(results_winter["Price"], results_winter["Success rate"], linewidth = 1.0)

x = 1002
y = 1.0

plt.scatter(x, y, color='#ff0000', zorder=1)

ax.set_ylabel("Success rate")
ax.set_xlabel("Price [â‚¬]")

plt.xlim([0,3000])

plt.savefig('results/results_winter.png')

plt.show()
```

```python
# Find optimal solution (summer)

i = results_summer.index
index = results_summer["Success rate"] > 0.99
optimal_result_s = i[index]
optimal_result_s.tolist()
print(optimal_result_s)

for i in optimal_result_s:
    display(results_summer.loc[[i]])
```

```python
# Find optimal solution (winter)

i = results_winter.index
index = results_winter["Success rate"] > 0.99
optimal_result_w = i[index]
optimal_result_w.tolist()
print(optimal_result_w)

for i in optimal_result_w:
    display(results_winter.loc[[i]])
```
