# Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
```

# Import summer / winter results

```python
# Import summer / winter results

with open("results/results_summer_highbatteryprice.csv", "r") as csvfile:
    results_summer = pd.read_csv(csvfile)

with open("results/results_winter_highbatteryprice.csv", "r") as csvfile:
    results_winter = pd.read_csv(csvfile)

results_summer = results_summer.sort_values("Price")

results_winter = results_winter.sort_values("Price")
```

# Plot price & success rate (summer)

```python
# Plot price / success rate (summer)

fig, ax = plt.subplots(figsize=(8,4))

plt.scatter(results_summer["Price"], results_summer["Success rate"], color = "b", alpha = 0.4)

x = 1990
y = 1.0

x1 = 1490
y1 = 0.950232

plt.scatter(x, y, color='r', zorder=1)
plt.scatter(x1, y1, color='r', zorder=1)

plt.plot([0,5000],[0.95,0.95], color = "black", linewidth = 1.0)

ax.set_ylabel("Success rate")
ax.set_xlabel("Price [€]")

plt.gca().invert_yaxis()

plt.xlim([0,5000])

legend_elements = [Line2D([0], [0], color='black', lw=1, label='Threshold'),
                   Line2D([0], [0], marker = "o", color='w', label='Optimal solutions',
                          markerfacecolor='r', markersize=10)]

ax.legend(handles=legend_elements, loc='upper right')

plt.savefig('results/results_summer.png')

plt.show()
```

# Plot price & success rate (winter)

```python
# Plot price / success rate (summer)

fig, ax = plt.subplots(figsize=(8,4))

plt.scatter(results_winter["Price"], results_winter["Success rate"], color = "b", alpha = 0.4)

x = 3490
y = 1.0

x2 = 2490
y2 = 0.952948

plt.scatter(x, y, color='r', zorder=1)
plt.scatter(x2, y2, color='r', zorder=1)

plt.plot([0,5000],[0.95,0.95], color = "black", linewidth = 1.0)

ax.set_ylabel("Success rate")
ax.set_xlabel("Price [€]")

plt.gca().invert_yaxis()

plt.xlim([0,5000])

legend_elements = [Line2D([0], [0], color='black', lw=1, label='Threshold'),
                   Line2D([0], [0], marker = "o", color='w', label='Optimal solutions',
                          markerfacecolor='r', markersize=10)]

ax.legend(handles=legend_elements, loc='upper right')

plt.savefig('results/results_winter.png')

plt.show()
```

# Winter / summer solar panel comparison

```python
# Winter / summer solar panel comparison

fig, ax = plt.subplots(figsize=(8,4))

scatter = plt.scatter(results_winter["Panel count"], results_winter["Success rate"], color = "b")

xticks = [1,2,3]

ax.set_ylabel("Success rate (Winter)")
ax.set_xlabel("Panel count")
ax.set_xticks(xticks)

plt.xlim([0.9,3.1])

plt.savefig('results/summer_winter_solarpanel1.png')

fig2, ax2 = plt.subplots(figsize=(8,4))

scatter2 = plt.scatter(results_summer["Panel count"], results_summer["Success rate"], color = "r")

ax2.set_ylabel("Success rate (Summer)")
ax2.set_xlabel("Panel count")
ax2.set_xticks(xticks)

plt.xlim([0.9,3.1])

plt.savefig('results/summer_winter_solarpanel2.png')

plt.show()

```

# Find optimal solution (summer)

```python
# Find optimal solution (summer)

i = results_summer.index
index = results_summer["Success rate"] > 0.95
optimal_result_s = i[index]
optimal_result_s.tolist()
print(optimal_result_s)

for i in optimal_result_s:
    display(results_summer.loc[[i]])
```
